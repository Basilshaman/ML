```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import accuracy_score,mean_squared_error
from sklearn.preprocessing import StandardScaler,LabelEncoder
from word2number import w2n
```

```python
data=pd.read_csv('Automobile_data.csv')
data
```

```python
data.isnull().sum()
```

```python
mean=data['price'].mean()
mean
```

```python
data['price']=data['price'].fillna(mean)

```

```python
data.dtypes
```

```python
pip install word2number
```

```python
lb=LabelEncoder()
```

```python
data['company']=lb.fit_transform(data['company'])
```

```python
data['body-style']=lb.fit_transform(data['body-style'])
```

```python
data['engine-type']=lb.fit_transform(data['engine-type'])
```

```python
data
```

```python
data.dtypes
```

```python
data['num-of-cylinders']=data['num-of-cylinders'].apply(w2n.word_to_num)
```

```python
data
```

```python
x=data.drop(['price'],axis=1)
y=data['price']
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=30)
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
svm=SVR(kernel='linear')
svm.fit(x_train,y_train)
svm.score(x_test,y_test)
```

```python
svm1=SVR(kernel='poly',degree=3,C=5,gamma='auto')
svm1.fit(x_train,y_train)
svm1.score(x_test,y_test)
```

```python
svm2=SVR(kernel='sigmoid')
svm2.fit(x_train,y_train)
svm2.score(x_test,y_test)
```

```python
svm3=SVR(kernel='rbf')
svm3.fit(x_train,y_train)
svm3.score(x_test,y_test)
```

```python
svm=SVR()
svm.fit(x_train,y_train)
svm.score(x_test,y_test)
```
