```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from word2number import w2n
```

```python
data=pd.read_csv('Automobile_data.csv')
data
```

```python
data.isnull().sum()
```

```python
mean=data['price'].mean()
mean
```

```python
data['price']=data['price'].fillna(mean)
```

```python
data.isnull().sum()
```

```python
lb=LabelEncoder()
```

```python
data['Acompany']=lb.fit_transform(data['company'])
```

```python
data['Abody_style']=lb.fit_transform(data['body-style'])
```

```python
data['Aengine_type']=lb.fit_transform(data['engine-type'])
```

```python
data['num-of-cylinders']=data['num-of-cylinders'].apply(w2n.word_to_num)
```

```python
data
```

```python
x=data.drop(['price','company','body-style','engine-type'],axis=1)
```

```python
x
```

```python
y=data['price']
y
```

```python
model=RandomForestRegressor(n_estimators=47,random_state=34)
model
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```
